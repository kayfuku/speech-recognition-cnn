```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import librosa
import librosa.display

def show_mfcc_NG(path, n_mfcc, max_len_time, sampling_rate):

    print(file_path)
    # duration: audio duration (sec)
#     wave, sr = librosa.load(file_path, sr=None, duration=5)
    wave, sr = librosa.load(file_path, duration=5)
    print('sr:', sr) # 22050
    print('len(wave):', len(wave)) # 30870

    mfcc = librosa.feature.mfcc(wave, sr=sampling_rate, n_mfcc=n_mfcc)
    print('mfcc.shape:', mfcc.shape)
    
    # Padding (time axis)
    if max_len_time > mfcc.shape[1]:
        pad_width = max_len_time - mfcc.shape[1]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad_width)), mode='constant')
    # Else cutoff the remaining parts
    else:
        mfcc = mfcc[:, :max_len_time]
    
    
#     mfcc = librosa.feature.mfcc(wave)
    print('mfcc.shape (after padding):', mfcc.shape)
#     print(mfcc)
    
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfcc, sr=sampling_rate, x_axis='time', y_axis='mel')
    plt.colorbar()
    plt.title('MFCC')
    plt.tight_layout()
    
    return mfcc
    
```

```python
#######################
# Log-mel spectrogram

def show_log_mel_NG(file_path, max_len_time):
    wave, sr = librosa.load(file_path, duration=5)

    mel_spect = librosa.feature.melspectrogram(y=wave, sr=16000, n_mels=128)
    
    # Convert a power spectrogram (amplitude squared) to decibel (dB) units
    # This computes the scaling 10 * log10(S / ref) in a numerically stable way. ref (default): 1
    mel_spect = librosa.power_to_db(mel_spect)
#     mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
    
    print('mel_spect.shape:', mel_spect.shape)
    
    # Padding (time axis)
    if max_len_time > mel_spect.shape[1]:
        pad_width = max_len_time - mel_spect.shape[1]
        mel_spect = np.pad(mel_spect, pad_width=((0, 0), (0, pad_width)), mode='constant')
    # Else cutoff the remaining parts
    else:
        mel_spect = mel_spect[:, :max_len_time]
    
    print('mel_spect.shape after padding:', mel_spect.shape)    
#     print(mel_spect)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')
    plt.title('Log Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    
    return mel_spect
```

```python
#######################
# Log-mel spectrogram

def show_log_mel2(file_path, max_len_time):
    wave, sr = librosa.load(file_path, sr=None, duration=5)

    mel_spect = librosa.feature.melspectrogram(y=wave, sr=sr, n_mels=128)
    
    # Convert a power spectrogram (amplitude squared) to decibel (dB) units
    # This computes the scaling 10 * log10(S / ref) in a numerically stable way. ref (default): 1
    mel_spect = librosa.power_to_db(mel_spect)
#     mel_spect = librosa.power_to_db(mel_spect, ref=np.max)
    
    print('mel_spect.shape:', mel_spect.shape)
    
    # Padding (time axis)
    if max_len_time > mel_spect.shape[1]:
        pad_width = max_len_time - mel_spect.shape[1]
        mel_spect = np.pad(mel_spect, pad_width=((0, 0), (0, pad_width)), mode='constant')
    # Else cutoff the remaining parts
    else:
        mel_spect = mel_spect[:, :max_len_time]
    
    print('mel_spect.shape after padding:', mel_spect.shape)    
#     print(mel_spect)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_spect, sr=sr, y_axis='mel', fmax=8000, x_axis='time')
    plt.title('Log Mel Spectrogram')
    plt.colorbar(format='%+2.0f dB')
    
    return mel_spect
```
